# 파이모디 첫 연결

##  핵심 개념
- **파이모디 구성**: 입력(6) + 출력(4) + 셋업(2)
- **LED 제어**: RGB 조합 (0~255)
- **Display**: 텍스트·숫자·변수 출력
- **f-string**: 변수값 포맷팅 (:.1f, :.0f)

##  파이모디 모듈 구성

### 출력 모듈 (Output)
| 모듈 | 기능 | 코드 |
|------|------|------|
| LED | RGB 색 빛 출력 (0~255) | `bundle.leds[0]` |
| Display | 텍스트·숫자 화면 출력 | `bundle.displays[0]` |
| Speaker | 음계·음량 소리 출력 | `bundle.speakers[0]` |
| Motor | 각도·속도 회전 운동 | `bundle.motors[0]` |

### 입력 모듈 (Input)
| 모듈 | 기능 | 코드 |
|------|------|------|
| Environment | 온도·습도·조도·소리 | `bundle.envs[0]` |
| Button | 클릭·더블클릭·누름 | `bundle.buttons[0]` |
| Dial | 회전 속도·각도 | `bundle.dials[0]` |
| ToF | 거리(근접도) 측정 | `bundle.tofs[0]` |
| IMU | 기울기·진동·가속도 | `bundle.imus[0]` |

### 셋업 모듈 (Setup)
- **Network**: 컴퓨터와 파이모디를 연결하는 통신 허브 (USB/BT)
- **Battery**: 무선 전원 공급

## 💡 RGB 색상 매핑표

| 색상 | R | G | B | 코드 |
|------|-----|-----|-----|------|
| 빨강 | 255 | 0 | 0 | `led.rgb = (255, 0, 0)` |
| 초록 | 0 | 255 | 0 | `led.rgb = (0, 255, 0)` |
| 파랑 | 0 | 0 | 255 | `led.rgb = (0, 0, 255)` |
| 노랑 | 255 | 255 | 0 | `led.rgb = (255, 255, 0)` |
| 보라 | 148 | 0 | 211 | `led.rgb = (148, 0, 211)` |
| 흰색 | 255 | 255 | 255 | `led.turn_on()` |
| 꺼짐 | 0 | 0 | 0 | `led.turn_off()` |

##  실습 A: 파이모디 연결 + LED 켜기



````python

import modi_plus, time

bundle = modi_plus.MODIPlus()
print(bundle.modules) # 연결 확인

led = bundle.leds[0]
led.rgb = (255, 0, 0) # 빨강
time.sleep(2)
led.rgb = (0, 255, 0) # 초록
time.sleep(2)
led.turn_off()
````

#  LED와 센서

## 핵심 개념
- **Button**: `clicked`, `double_clicked`, `pressed`, `toggled` 4가지 상태
- **Environment 센서**: 온도, 습도, 조도(빛), 소리
- **while True 루프**: 실시간 센서값 모니터링
- **조건문**: if-elif-else로 상태별 LED 제어

##  Button 4가지 상태

| 상태 | 설명 | 코드 |
|------|------|------|
| clicked | 버튼 한 번 클릭 | `button.clicked` |
| double_clicked | 버튼 두 번 클릭 | `button.double_clicked` |
| pressed | 누르고 있는 상태 | `button.pressed` |
| toggled | ON/OFF 전환 | `button.toggled` |

## Environment 센서값

| 속성 | 측정 대상 | 범위 | 코드 |
|------|---------|------|------|
| illuminance | 빛의 밝기 (조도) | 0~100% | `env.illuminance` |
| volume | 주변 소리 크기 | 0~100% | `env.volume` |
| temperature | 온도 (°C) | -40~125 | `env.temperature` |
| humidity | 습도 (%) | 0~100 | `env.humidity` |

##  실습 A: 페이퍼 무드등



````python

import modi_plus, time

bundle = modi_plus.MODIPlus()
button = bundle.buttons[0]
led = bundle.leds[0]

colors = [
    (0, 0, 0),       # 꺼짐
    (255, 0, 0),     # 빨강
    (0, 255, 0),     # 초록
    (0, 0, 255),     # 파랑
]
mode = 0

while True:
    if button.double_clicked:
        led.turn_off()
        break
    if button.clicked:
        mode = (mode + 1) % 4
        led.rgb = colors[mode]
    time.sleep(0.1)
````

# Display 활용하기

##  핵심 개념
- **Display 명령어**: `.text`, `.reset()`, f-string
- **포맷 지정**: `:.1f` (소수 1자리), `:.0f` (정수)
- **실시간 표시**: while True + time.sleep()
- **Dial 모듈**: 0~100 범위 회전값

## Display 명령어

| 명령어 | 설명 | 예시 |
|--------|------|------|
| `.text = "..."` | 텍스트 출력 (한 줄) | `display.text = "Hello!"` |
| `.reset()` | 화면 전체 지우기 | `display.reset()` |
| f-string | 변수+텍스트 조합 | `display.text = f"T:{temp}C"` |
| `:.1f` | 소수점 1자리 | `f"{val:.1f}"` → "25.3" |
| `:.0f` | 소수점 없이 정수 | `f"{val:.0f}"` → "25" |

# 다중 센서 조합

##  핵심 개념
- **and**: 둘 다 참이어야 True (긴급 상황)
- **or**: 하나만 참이면 True
- **우선순위 로직**: 긴급한 조건을 먼저 검사 (if → elif → else)
- **매핑표 설계**: 조건을 표로 정리 → AI가 정확한 코드 생성

##  and/or/not 정리

| 연산자 | 의미 | 예시 | 결과 |
|--------|------|------|------|
| and | 둘 다 참이어야 True | 빛<30 and 소리>50 | 둘 다 맞아야 실행 |
| or | 하나만 참이면 True | 온도>35 or 습도>80 | 하나만 맞아도 실행 |
| not | 반대로 뒤집기 | not button.clicked | True→False 변환 |

##  5단계 우선순위 매핑표 예시

| 순위 | 조건 | LED | Display | 의미 |
|------|------|-----|---------|------|
| 1 (최고) | 빛<30 AND 소리>50 | 빨강 깜빡 | EMERGENCY! | 어둡고 시끄러운 긴급 |
| 2 | 온도>35 | 빨강 | HOT! | 고온 경고 |
| 3 | 빛<30 | 노랑 | DARK | 어두움 알림 |
| 4 | 소리>50 | 파랑 깜빡 | LOUD | 소음 알림 |
| 기본 | 그 외 | 초록 | ALL OK | 정상 상태 |

## 기본 예제: 5단계 경보 시스템

```python
import modi_plus, time

bundle = modi_plus.MODIPlus()
env = bundle.envs[0]
led = bundle.leds[0]
display = bundle.displays[0]

def blink(color, n=3):
    for _ in range(n):
        led.rgb = color
        time.sleep(0.2)
        led.turn_off()
        time.sleep(0.2)

while True:
    li = env.illuminance
    so = env.volume
    te = env.temperature
    
    # 1순위: 가장 긴급한 조건 (AND)
    if li < 30 and so > 50:
        blink((255, 0, 0))
        display.text = "EMERGENCY!"
    # 2순위
    elif te > 35:
        led.rgb = (255, 0, 0)
        display.text = f"HOT!{te:.0f}C"
    # 3순위
    elif li < 30:
        led.rgb = (255, 255, 0)
        display.text = "DARK"
    # 4순위
    elif so > 50:
        blink((0, 0, 255))
        display.text = "LOUD"
    # 기본
    else:
        led.rgb = (0, 255, 0)
        display.text = "ALL OK"
    
    time.sleep(0.5)
```

```python
import modi_plus, time

bundle = modi_plus.MODIPlus()
env = bundle.envs[0]
led = bundle.leds[0]
display = bundle.displays[0]

# 🔧 ====== 매핑표 기준값을 직접 수정해보세요! ======
빛기준 = 30     # 이 값보다 어두우면 DARK
소리기준 = 50   # 이 값보다 크면 LOUD
온도기준 = 35   # 이 값보다 높으면 HOT
체크주기 = 0.5
# =====================================================

def blink(color, n=3):
    for _ in range(n):
        led.rgb = color
        time.sleep(0.2)
        led.turn_off()
        time.sleep(0.2)

while True:
    li = env.illuminance
    so = env.volume
    te = env.temperature

    if li < 빛기준 and so > 소리기준:
        blink((255, 0, 0))
        display.text = "EMERGENCY!"
    elif te > 온도기준:
        led.rgb = (255, 0, 0)
        display.text = f"HOT!{te:.0f}C"
    elif li < 빛기준:
        led.rgb = (255, 255, 0)
        display.text = "DARK"
    elif so > 소리기준:
        blink((0, 0, 255))
        display.text = "LOUD"
    else:
        led.rgb = (0, 255, 0)
        display.text = "ALL OK"

    time.sleep(체크주기)
```

- 기준값(`빛기준`, `소리기준`, `온도기준`) 숫자만 바꿔서 우리 환경에 맞춰보기
- `blink()`의 `n` 값을 바꿔서 깜빡이는 횟수 조절하기

# 무선통신으로 파이모디 연결

## 핵심 개념
- **BLE (Bluetooth Low Energy)**: 무선 연결 방식
- **connection_type="ble"**: 무선 모드 설정
- **network_uuid**: 파이모디 고유 식별 번호
- **코드 변경점**: USB는 `connection_type="ble"`, **나머지는 동일!**

##  유선 vs 무선 비교

| 항목 | USB 유선 | Bluetooth 무선 | Wi-Fi 무선 |
|------|---------|---------------|----------|
| 연결 방식 | USB 케이블 | Bluetooth 페어링 | Wi-Fi 네트워크 |
| 거리 | ~2m (케이블 길이) | ~10m | ~50m (실내) |
| 속도 | 빠름 | 보통 | 빠름 |
| 장점 | 안정적, 전원 공급 | 간편, 1:1 연결 | 여러 기기 연결 |

| 활용 | 초기 테스트 | 로봇, 자동차 제어 | 원격 모니터링 |

##  BLE 연결 순서

| 단계 | 작업 | 설명 |
|------|------|------|
| 1 | Network 모듈 전원 ON | USB 분리 후 배터리/외부 전원 연결 |
| 2 | BLE 모드 확인 | Network 모듈 LED가 파란색으로 깜빡이면 BLE 대기 상태 |
| 3 | 컴퓨터 블루투스 ON | 설정 → Bluetooth → 켜기 |
| 4 | pymodi 코드에서 연결 | `MODIPlus(connection_type="ble")` 로 BLE 연결 |
| 5 | 연결 확인 | `bundle.modules` 로 모듈 목록 출력 |

# 키보드로 자동차 운전하기

## 핵심 개념
- **motor.speed**: -100~100 (양수=정회전, 음수=역회전, 0=정지)
- **motor.angle**: 목표 각도로 회전 (0~360)
- **pynput**: 키보드 이벤트 감지
- **on_press & on_release**: 키 누름/뗌 시 자동 실행

## 모터 제어 명령어

| 명령어 | 설명 | 범위 | 예시 |
|--------|------|------|------|
| .speed | 회전 속도 설정 | -100~100 | `motor.speed = 50` |
| .angle | 목표 각도로 회전 | 0~360 | `motor.angle = 90` |
| 양수/음수 | 회전 방향 결정 | + 정회전, - 역회전 | `motor.speed = -50` |
| 0 | 정지 | 속도 0 = 멈춤 | `motor.speed = 0` |

## 자동차 모터 제어 원리

| 동작 | 왼쪽 모터 | 오른쪽 모터 | 방향 |
|------|---------|---------|------|
| 직진 | 전진(+) | 전진(+) | ⬆ 앞으로 |
| 후진 | 역방향(-) | 역방향(-) | ⬇ 뒤로 |
| 좌회전 | 정지(0) | 전진(+) | ⬅ 왼쪽 |
| 우회전 | 전진(+) | 정지(0) | ➡ 오른쪽 |
| 정지 | 정지(0) | 정지(0) | ⏹ 멈춤 |

## 기본 예제: 모터 기본 제어

```python
import modi_plus, time

bundle = modi_plus.MODIPlus()
motor = bundle.motors[0]

motor.speed = 50  # 정회전 50%
time.sleep(2)
motor.speed = -50 # 역회전 50%
time.sleep(2)
motor.speed = 0   # 정지
```

## 실습 A: 전진 → 우회전 → 전진

```python
import modi_plus, time

bundle = modi_plus.MODIPlus()
m_left = bundle.motors[0]
m_right = bundle.motors[1]

def forward(sec):
    m_left.speed = 50
    m_right.speed = 50
    time.sleep(sec)

def turn_right(sec):
    m_left.speed = 50
    m_right.speed = 0
    time.sleep(sec)

def stop():
    m_left.speed = 0
    m_right.speed = 0

time.sleep(2)
forward(2)
turn_right(1)
forward(2)
stop()
```

**핵심 개념**:
- `on_press`: 키를 누를 때 호출되는 함수
- `on_release`: 키를 뗄 때 호출되는 함수
- `Listener`: 키 이벤트를 감지하는 리스너

## 키보드 매핑 설계

| 키 | 방향 | 왼쪽 모터 | 오른쪽 모터 | 동작 |
|-------|------|---------|---------|------|
| W | 전진 | speed = 50 | speed = 50 | ⬆ |
| S | 후진 | speed = -50 | speed = -50 | ⬇ |
| A | 좌회전 | speed = 0 | speed = 50 | ⬅ |
| D | 우회전 | speed = 50 | speed = 0 | ➡ |
| Space | 정지 | speed = 0 | speed = 0 | ⏹ |
| Q | 종료 | - | - | 프로그램 종료 |

## 기본 예제: WASD 키보드 자동차 제어

```python
import modi_plus
from pynput import keyboard

bundle = modi_plus.MODIPlus()
m_l = bundle.motors[0]  # 왼쪽
m_r = bundle.motors[1]  # 오른쪽
led = bundle.leds[0]
display = bundle.displays[0]

SPD = 60  # 기본 속도

def go(l, r, msg, color):
    m_l.speed = l
    m_r.speed = r
    display.text = msg
    led.rgb = color

def stop():
    go(0, 0, "STOP", (0,0,0))

def on_press(key):
    try:
        k = key.char
        if k == "w":
            go(SPD, SPD, "FORWARD", (0,255,0))
        elif k == "s":
            go(-SPD, -SPD, "BACK", (255,0,0))
        elif k == "a":
            go(0, SPD, "LEFT", (0,0,255))
        elif k == "d":
            go(SPD, 0, "RIGHT", (255,255,0))
        elif k == "q":
            return False  # 종료
    except: pass

def on_release(key):
    stop()  # 키 뗄 때 정지

print("WASD=이동 Q=종료")
with keyboard.Listener(on_press=on_press, on_release=on_release) as listener:
    listener.join()
```
