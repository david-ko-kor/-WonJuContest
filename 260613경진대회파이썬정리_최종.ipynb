## 반환값 (return)

```python
# 반환값이 있는 함수
def multiply(a, b):
    result = a * b
    return result

# 반환값을 변수에 저장
x = multiply(10, 5)
y = multiply(3, 4)
print(f"x = {x}")
print(f"y = {y}")
print(f"multiply(7, 8) = {multiply(7, 8)}")
```

## 함수 활용: 학점 판정

```python
def get_grade(score):
    if score >= 90:
        return "A"
    elif score >= 80:
        return "B"
    elif score >= 70:
        return "C"
    else:
        return "F"

# 함수 사용
print(f"get_grade(95) = {get_grade(95)}")
print(f"get_grade(85) = {get_grade(85)}")
print(f"get_grade(65) = {get_grade(65)}")
```

##  함수 활용: 나이 계산

```python
def calculate_age(birth_year):
    current_year = 2026
    age = current_year - birth_year
    return age

def is_student_status(age):
    if age >= 14:
        return "중학생"
    else:
        return "초등학생"

age = calculate_age(2012)
status = is_student_status(age)
print(f"나이: {age}살, 상태: {status}")
```

## 함수 안에 조건문

```python
def check_pass(score):
    if score >= 70:
        return "합격"
    else:
        return "불합격"

def get_discount(age):
    if age <= 13:
        return 0.3  # 30% 할인
    elif age >= 65:
        return 0.2  # 20% 할인
    else:
        return 0    # 할인 없음

print(f"check_pass(75) = {check_pass(75)}")
print(f"get_discount(10) = {get_discount(10)}")
print(f"get_discount(30) = {get_discount(30)}")
print(f"get_discount(70) = {get_discount(70)}")
```

# # 파이썬 기초 실습 6: while문

## while문이란?
- **while문**: 조건이 참인 동안 반복
- **조건**: True/False로 나타냄
- **중요**: 반복 중에 조건을 변경해야 함!

---

```python
count = 1  # 시작값

while count <= 5:
    print(f"반복 {count}번")
    count = count + 1
```

## 예제 2: 조건을 확인하면서 반복하기

조건이 언제 False가 되는지 관찰하세요

```python
num = 1

while num <= 3:
    print(f"현재 num={num}, 조건 (num <= 3)은 {num <= 3}")
    num = num + 1

print(f"루프 종료 후 num={num}, 조건 (num <= 3)은 {num <= 3}")
```

## 예제 3: 카운트다운

5부터 1까지

```python
countdown = 5

while countdown > 0:
    print(f"{countdown}")
    countdown = countdown - 1

print("발사! 🚀")
```

## 예제 4: 1부터 10까지 합 계산하기

```python
total = 0
i = 1

while i <= 10:
    total = total + i
    print(f"i={i}, 누적합={total}")
    i = i + 1

print(f"\n최종 합계: {total}")
```

# 파이썬 기초 실습 5: for문

## for문이란?
- **반복문**: 같은 작업을 여러 번 하기
- **for문**: 정해진 횟수만큼 반복
- **range()**: 범위를 지정하는 함수

---

## 예제 1: 가장 간단한 for문

5번 반복하기

```python
# 5번 반복
for i in range(5):
    print(f"반복 {i+1}번")
```

## 예제 2: 반복 변수의 값 확인

i의 값이 0부터 4까지 변합니다

```python
print("i의 값이 어떻게 변하는지 보세요:")
for i in range(5):
    print(f"현재 i = {i}")
```

## 예제 3: 1부터 5까지 출력하기

range(1, 6)을 사용합니다

```python
for i in range(1, 6):
    print(f"번호: {i}")

print("\n💡 range(1, 6)은 1부터 5까지 (6은 포함 안 함)")
```
